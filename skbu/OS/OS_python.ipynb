# CPU scheduling algorithm

## 1. FCFS 

```python
def fcfs(processes):
  processes.sort(key=lambda x: x["AT"])
  # print(processes)
  PST = 0  # Process Start Time
  total_BT=0
  total_TAT=0
  total_CT=0
  total_WT=0
  total_RT=0
  list_of_CT = []
  list_of_AT = []
  
  print("PID\tAT\tBT\tCT\tTAT\tWT\tRT")
  
  x=0
  for process in processes:
    pid = process["pid"]
    AT = process["AT"]
    BT = process["BT"]
    
    total_BT = total_BT + BT
    
    # if cpu idle or not
    if PST < AT:
      PST = AT
      
    # completion time
    CT = PST + BT
    total_CT = total_CT + CT
    # store all CT & AT in a list
    list_of_CT.append(CT)
    list_of_AT.append(AT)
    
    
    # turn around time
    TAT = CT - AT
    total_TAT = total_TAT + TAT
    
    # waiting time
    WT = CT - AT - BT
    total_WT = total_WT + WT
    
    # response time
    RT = PST - AT
    total_RT = total_RT + RT
    
    # Updation of PST
    PST = PST + BT
    
    print(f"p{pid}\t{AT}\t{BT}\t{CT}\t{TAT}\t{WT}\t{RT}")
  
  print('\n---------------------------------------------------\n')
  avg_TAT = total_TAT/len(processes)
  avg_WT = total_WT/len(processes)
  avg_RT = total_RT/len(processes)
  scheduling_length = max(list_of_CT) - min(list_of_AT)
  throughput = len(processes)/scheduling_length
  CPU_utilization = total_BT/PST*100
  
  print(f"Avarage Turn-Around-Time: {avg_TAT} units")
  print(f"Avarage Waiting-Time: {avg_WT} units")
  print(f"Avarage Response-Time: {avg_RT} units")
  print(f"Avarage scheduling_length: {scheduling_length} units")
  print(f"Avarage throughput: {throughput:.2f} process/unit")
  print(f"Avarage CPU_utilization: {CPU_utilization}%")
  print('\n---------------------------------------------------\n')
  
  

# processes = [
#   {"pid":1, "AT":1, "BT":3},
#   {"pid":2, "AT":1, "BT":2},
#   {"pid":3, "AT":2, "BT":1},
#   {"pid":4, "AT":3, "BT":4},
#   {"pid":5, "AT":4, "BT":5},
# ]
processes = [
  {"pid":1, "AT":0, "BT":3},
  {"pid":2, "AT":0, "BT":2},
  {"pid":3, "AT":0, "BT":1},
  {"pid":4, "AT":0, "BT":4},
  {"pid":5, "AT":0, "BT":5},
]
fcfs(processes)
```

## 2. SJF

## 3. SRTF

```python
class Process:
    def __init__(self, id, arrival_time, burst_time):
        self.id = id
        self.AT = arrival_time  # Arrival Time
        self.BT = burst_time  # Burst Time
        self.remaining_time = burst_time  # Remaining execution time
        self.CT = 0  # Completion Time
        self.TAT = 0  # Turn Around Time
        self.WT = 0  # Waiting Time
        self.RT = None  # Response Time

def srtf(processes):
    current_time, completed = 0, 0
    gantt_chart = []
    n = len(processes)
    processes.sort(key=lambda x: x.AT)  # Sort by Arrival Time
    
    while completed < n:
        available_processes = [p for p in processes if p.AT <= current_time and p.remaining_time > 0]
        
        if available_processes:
            shortest = min(available_processes, key=lambda x: x.remaining_time)
            
            if shortest.RT is None:  
                shortest.RT = current_time - shortest.AT  # First time the process gets CPU
            
            shortest.remaining_time -= 1
            current_time += 1
            gantt_chart.append(f"P{shortest.id}({current_time})")
            
            if shortest.remaining_time == 0:
                shortest.CT = current_time
                shortest.TAT = shortest.CT - shortest.AT
                shortest.WT = shortest.TAT - shortest.BT
                completed += 1
        else:
            gantt_chart.append(f"Idle({current_time+1})")
            current_time += 1  # No process was available, so CPU remains idle

    return gantt_chart, current_time

def print_results(processes, gantt_chart, scheduling_length):
    total_WT, total_TAT, total_RT = 0, 0, 0
    
    print("\nPID\tAT\tBT\tCT\tTAT\tWT\tRT")
    for p in processes:
        print(f"{p.id}\t{p.AT}\t{p.BT}\t{p.CT}\t{p.TAT}\t{p.WT}\t{p.RT}")
        total_WT += p.WT
        total_TAT += p.TAT
        total_RT += p.RT

    n = len(processes)
    avg_WT = total_WT / n
    avg_TAT = total_TAT / n
    avg_RT = total_RT / n
    
    throughput = n / scheduling_length  # Throughput calculation
    cpu_utilization = (scheduling_length - gantt_chart.count("Idle")) / scheduling_length * 100  # CPU Utilization
    
    print("\n---------------------------------------------------\n")
    print(f"Average Turn-Around Time: {avg_TAT:.2f} units")
    print(f"Average Waiting Time: {avg_WT:.2f} units")
    print(f"Average Response Time: {avg_RT:.2f} units")
    print(f"Scheduling Length: {scheduling_length} units")
    print(f"Throughput: {throughput:.2f} processes/unit")
    print(f"CPU Utilization: {cpu_utilization:.2f}%")
    print("\nGantt Chart:")
    print(" | ".join(gantt_chart))
    print("\n---------------------------------------------------\n")

# Process list data
process_list = [
    {"pid": 1, "AT": 0, "BT": 7},
    {"pid": 2, "AT": 1, "BT": 5},
    {"pid": 3, "AT": 2, "BT": 3},
    {"pid": 4, "AT": 3, "BT": 1},
    {"pid": 5, "AT": 4, "BT": 2},
    {"pid": 6, "AT": 5, "BT": 1},
]

# Convert dictionary data to Process objects
processes = [Process(p["pid"], p["AT"], p["BT"]) for p in process_list]
gantt_chart, scheduling_length = srtf(processes)
print_results(processes, gantt_chart, scheduling_length)
```

## 4. Round Robin

```python
def round_robin(processes, quantum):
    time = 0
    queue = []
    gantt_chart = []
    list_of_CT = []
    list_of_AT = [p["AT"] for p in processes]

    total_BT = sum(p["BT"] for p in processes)
    for p in processes:
        p["remaining_time"] = p["BT"]
        p["CT"] = p["TAT"] = p["WT"] = 0
        p["RT"] = -1  # Initially set to -1 for response time tracking

    processes.sort(key=lambda x: x["AT"])
    queue.append(processes[0])  # Start with the first process

    index = 1  # Tracks processes arriving later
    while queue:
        process = queue.pop(0)

        if process["RT"] == -1:  
            process["RT"] = time - process["AT"]

        execution_time = min(process["remaining_time"], quantum)
        process["remaining_time"] -= execution_time
        time += execution_time
        gantt_chart.append(f"P{process['pid']} ({time})")

        # Add new arriving processes
        while index < len(processes) and processes[index]["AT"] <= time:
            queue.append(processes[index])
            index += 1

        if process["remaining_time"] > 0:  
            queue.append(process)
        else:
            process["CT"] = time
            process["TAT"] = process["CT"] - process["AT"]
            process["WT"] = process["TAT"] - process["BT"]
            list_of_CT.append(process["CT"])

    print("PID\tAT\tBT\tCT\tTAT\tWT\tRT")
    total_tat = total_wt = total_rt = 0
    for p in processes:
        print(f"{p['pid']}\t{p['AT']}\t{p['BT']}\t{p['CT']}\t{p['TAT']}\t{p['WT']}\t{p['RT']}")
        total_tat += p["TAT"]
        total_wt += p["WT"]
        total_rt += p["RT"]

    avg_tat = total_tat / len(processes)
    avg_wt = total_wt / len(processes)
    avg_rt = total_rt / len(processes)
    scheduling_length = max(list_of_CT) - min(list_of_AT)
    throughput = len(processes) / scheduling_length
    CPU_utilization = (total_BT / time) * 100

    print("\n---------------------------------------------------\n")
    print(f"Average Turn-Around-Time: {avg_tat:.2f} units")
    print(f"Average Waiting-Time: {avg_wt:.2f} units")
    print(f"Average Response-Time: {avg_rt:.2f} units")
    print(f"Scheduling Length: {scheduling_length} units")
    print(f"Throughput: {throughput:.2f} processes/unit")
    print(f"CPU Utilization: {CPU_utilization:.2f}%")
    print("\n---------------------------------------------------\n")

    print("Gantt Chart:")
    print(" | ".join(gantt_chart))


# Example Usage
process_list = [
    {"pid": 1, "AT": 0, "BT": 5},
    {"pid": 2, "AT": 1, "BT": 3},
    {"pid": 3, "AT": 2, "BT": 1},
    {"pid": 4, "AT": 3, "BT": 2},
    {"pid": 5, "AT": 4, "BT": 3},
]

time_quantum = 2  # Example time quantum
round_robin(process_list, time_quantum)
```

## 5. HRRN

```python
class Process:
    def __init__(self, id, arrival_time, burst_time):
        self.id = id
        self.AT = arrival_time  # Arrival Time
        self.BT = burst_time  # Burst Time
        self.CT = 0  # Completion Time
        self.TAT = 0  # Turn Around Time
        self.WT = 0  # Waiting Time
        self.RR = 0  # Response Ratio

def hrrn(processes):
    current_time = 0
    completed = 0
    gantt_chart = []
    n = len(processes)
    processes.sort(key=lambda x: x.AT)  # Sort by Arrival Time
    
    while completed < n:
        available_processes = [p for p in processes if p.AT <= current_time and p.CT == 0]
        
        if available_processes:
            for p in available_processes:
                p.RR = (current_time - p.AT + p.BT) / p.BT  # Calculate Response Ratio
            highest_rr_process = max(available_processes, key=lambda x: x.RR)
            
            gantt_chart.append(f"P{highest_rr_process.id}({current_time + highest_rr_process.BT})")
            current_time += highest_rr_process.BT
            highest_rr_process.CT = current_time
            highest_rr_process.TAT = highest_rr_process.CT - highest_rr_process.AT
            highest_rr_process.WT = highest_rr_process.TAT - highest_rr_process.BT
            completed += 1
        else:
            gantt_chart.append(f"Idle({current_time + 1})")
            current_time += 1  # No process was available, so CPU remains idle

    return gantt_chart, current_time

def print_results(processes, gantt_chart, scheduling_length):
    total_WT, total_TAT = 0, 0
    
    print("\nPID\tAT\tBT\tCT\tTAT\tWT")
    for p in processes:
        print(f"{p.id}\t{p.AT}\t{p.BT}\t{p.CT}\t{p.TAT}\t{p.WT}")
        total_WT += p.WT
        total_TAT += p.TAT

    n = len(processes)
    avg_WT = total_WT / n
    avg_TAT = total_TAT / n
    
    throughput = n / scheduling_length  # Throughput calculation
    cpu_utilization = (scheduling_length - gantt_chart.count("Idle")) / scheduling_length * 100  # CPU Utilization
    
    print("\n---------------------------------------------------\n")
    print(f"Average Turn-Around Time: {avg_TAT:.2f} units")
    print(f"Average Waiting Time: {avg_WT:.2f} units")
    print(f"Scheduling Length: {scheduling_length} units")
    print(f"Throughput: {throughput:.2f} processes/unit")
    print(f"CPU Utilization: {cpu_utilization:.2f}%")
    print("\nGantt Chart:")
    print(" | ".join(gantt_chart))
    print("\n---------------------------------------------------\n")

# Process list data
process_list = [
    {"pid": 1, "AT": 0, "BT": 5},
    {"pid": 2, "AT": 1, "BT": 3},
    {"pid": 3, "AT": 2, "BT": 8},
    {"pid": 4, "AT": 3, "BT": 6},
]

# Convert dictionary data to Process objects
processes = [Process(p["pid"], p["AT"], p["BT"]) for p in process_list]
gantt_chart, scheduling_length = hrrn(processes)
print_results(processes, gantt_chart, scheduling_length)
```

# Page replacement policy

## 1. FIFO

```python
def fifo_page_replacement(pages,frame_count): 
  frame=[]
  index=0
  page_fault = 0 #store the no of page fault
  print("Page\tFrames\t\tPage Fault")
  for page in pages:
    if page not in frame:
        page_fault += 1
        if len(frame) < frame_count:
            frame.append(page)
        else:
            frame[index] = page
            index = (index + 1) % frame_count
        print(f"{page}\t{frame}\tYes")
    else:
        print(f"{page}\t{frame}\tNo")
  print(f"No. of pagefault: {page_fault}")
  no_of_hit = len(pages)  - page_fault
  hit_ratio = (no_of_hit/len(pages)) * 100
  print(f"hit ratio: {hit_ratio}")

pages = [1,2,3,4,1,2,5,1,2,3,4,5]
frame_count = 3
fifo_page_replacement(pages, frame_count)
```

## 2. LRU

```python
def lru_page_replacement(pages, frame_count):
  frame  = []
  # index = 0
  page_fault = 0 #store the no of page fault
  recent_indices = {} 
  print("Page\tFrames\t\tPage Fault")
  for index in range(len(pages)):
      page = pages[index]
      if page not in frame:
          page_fault += 1
          if len(frame) < frame_count:
              frame.append(page)
          else:
              lru_page = min(recent_indices,key=recent_indices.get)
              frame[frame.index(lru_page)] = page
              del recent_indices[lru_page]
          falut_status = "YES"
      else:
          falut_status = "NO"
      recent_indices[page] = index
      print(f"{page}\t{frame}\t{falut_status}")
  print(f"no.of pagefault: {page_fault}")
  no_of_hit = len(pages)  - page_fault
  hit_ratio = (no_of_hit/len(pages)) * 100
  print(f"hit ratio: {hit_ratio:.2f}")

pages = [1,2,3,4,1,2,5,1,2,3,4,5]
frame_count = 3
lru_page_replacement(pages, frame_count)
```

## 3. Optimal

```python
def optimal_page_replacement(pages, frame_count):
  frames = []  # To store pages in memory
  page_faults = 0
  
  print("Page\tFrames\tPage Fault")
  
  for i in range(len(pages)):
      if pages[i] in frames:
          print(f"{pages[i]}\t{frames}\tNo")
          continue  # Page hit, no replacement needed
      
      if len(frames) < frame_count:
          frames.append(pages[i])  # If space is available, add the page
      else:
          # Find the page to replace
          farthest_index = -1
          page_to_replace = None
          
          for page in frames:
              if page not in pages[i+1:]:  # If the page is not used in the future, replace it
                  page_to_replace = page
                  break
              else:
                  index = pages[i+1:].index(page)  # Find the next occurrence
                  if index > farthest_index:
                      farthest_index = index
                      page_to_replace = page
          
          # Replace the page
          frames[frames.index(page_to_replace)] = pages[i]
      
      page_faults += 1  # Every replacement or addition causes a page fault
      print(f"{pages[i]}\t{frames}\tYes")
  
  print("Total Page Faults:", page_faults)

# Example usage
pages = [1,2,3,4,1,2,5,1,2,3,4,5]
frame_count = 3
optimal_page_replacement(pages, frame_count)
```
